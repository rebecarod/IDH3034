```python
# Dictionary that maps states to their abbreviations.
states = {
    'Oregon': 'OR',
    'Florida': 'FL',
    'California': 'CA',
    'New York': 'NY',
    'Michigan': 'MI'
}

# Dictionaries that maps cities to the abbreviations of states
cities = {
    'CA': 'San Francisco',
    'MI': 'Detroit',
    'FL': 'Jacksonville'
}

# Add to the dictionary called 'cities' two new cities to the abbreviations of states.
cities['NY'] = 'New York'
cities['OR'] = 'Portland'
print('-' * 10)

# Pull the city mapped to NY in the cities dictionary
print('NY state has: ', cities['NY'])

# Pull the city mapped to OR in the cities dictionary
print('OR state has: ', cities['OR'])

print('-' * 10)
# call the states dictionary and pull the abbreviation mapped to 'michigan'
print('Michigan\'s abbreviation is: ', states["Michigan"])
# call the states dictionary and pull the abbreviation mapped to 'florida'
print('Florida\'s abbreviation is: ', states['Florida'])
print('-' * 10)
# Pull the states dictionary the abbreviation linked to 'Michigan', then in the cities dictionary pull the city linked to 'MI'
print('Michigan has: ', cities[states['Michigan']])
# call the states dictionary and pull the abbreviation mapped to 'Florida', then in the cities dictionary pull the city linked to 'FL'
print('Florida has: ', cities[states['Florida']])
print('-' * 10)
# Loop through all the states dictionary
for state, abbrev in list(states.items()):
    # pull first value (state) then : second value (abbreviation)
    print(f"{state} is abbreviated {abbrev}")
    
print('-' * 10)
# Loop through the cities dictionary
for abbrev, city in list(cities.items()):
    # Pull the first value (abbreviation) : then the second value (cities)
    print(f"{abbrev} has the city {city}.")

print('-' * 10)
# Loop through the states dictionary
for state, abbrev in list(states.items()):
    # Pull the states and abbreviation from states dictionary
    print(f"{state} is abbreviated {abbrev}.")
    # Pull the abbreviation from the cities dictionary
    print(f"and has cities {cities[abbrev]}\n")

print('-' * 10)
# Pull value 'Texas' from the dictionary states
state = states.get('Texas')
# If Texas DNE in states dictionary print out a message
if not state:
    print("Sorry, no Texas.")

# Get TX from cities dictionary with 'Does not exist' as the city.
city = cities.get('TX', 'Does Not Exist')
print(f"The city for the state 'TX' is: {city}")
```
