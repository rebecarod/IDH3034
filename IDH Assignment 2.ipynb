Question 3: The column 'name' has the last name and the first name separated by a comma. Replace the column 'name' with 2 columns 'last_name' and 'first_name'. The exact method for doing this has not been taught in class. Feel free to use google and stackoverflow to figure out a way.

```python
newname = titanic["name"].str.split(",",n = 1,expand = True)

titanic["Last Name"] = newname[0]
titanic["First Name"] = newname[1]
titanic.drop(columns = ["name"], inplace = True)
titanic.head()
```
